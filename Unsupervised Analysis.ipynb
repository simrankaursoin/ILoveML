```python
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split 
import numpy as np
import math
import pandas as pd
pd.options.mode.chained_assignment = None

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
drug_df = pd.read_csv("drug200.csv")
drug_df.head()
```

```python
# drug type is the target feature
x = drug_df[["Age", "Sex", "BP", "Cholesterol", "Na_to_K"]]
y = drug_df[["Drug"]]
```

<h1>Visualization</h1>

```python
plt.scatter(y["Drug"], x["Age"], color="blue")
plt.ylabel("Age")
plt.show()
plt.scatter(y["Drug"], x["Sex"], color="red")
plt.ylabel("Sex")
plt.show()
plt.scatter(y["Drug"], x["BP"], color="green")
plt.ylabel("BP")
plt.show()
plt.scatter(y["Drug"], x["Cholesterol"], color="orange")
plt.ylabel("Cholesterol")
plt.show()
plt.scatter(y["Drug"], x["Na_to_K"], color="purple")
plt.ylabel("Na_to_K")
plt.show()
```

```python
plt.hist(x["Age"], color="blue")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
plt.hist(x["Sex"], color="red")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
plt.hist(x["BP"], color="green")
plt.xlabel("BP")
plt.ylabel("Count")
plt.show()
plt.hist(x["Cholesterol"], color="orange")
plt.xlabel("Cholesterol")
plt.ylabel("Count")
plt.show()
plt.hist(x["Na_to_K"], color="purple")
plt.xlabel("Na_to_K")
plt.ylabel("Count")
plt.show()
```

<h1>Clustering</h1>

```python
x["Sex"] = x["Sex"].map({"F":0, "M":1})
x["BP"] = x["BP"].map({"LOW":0, "NORMAL":1, "HIGH":2})
x["Cholesterol"] = x["Cholesterol"].map({"LOW":0, "NORMAL":1, "HIGH":2})
y["Drug"] = y["Drug"].map({"drugA":0, "drugB":1, "drugC":2, "drugX":3, "DrugY":4})
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(x[["BP", "Age"]])
y_kmeans = kmeans.predict(x[["BP", "Age"]])
print("Colors = predicted clusters")
plt.scatter(x["BP"], x["Age"], c=y_kmeans)
plt.xlabel("BP")
plt.ylabel("Age")
plt.show()
print("Colors = actual clusters")
plt.scatter(x["BP"], x["Age"], c=y["Drug"])
plt.xlabel("BP")
plt.ylabel("Age")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(x[["Cholesterol", "Age"]])
y_kmeans = kmeans.predict(x[["Cholesterol", "Age"]])
print("Colors = predicted clusters")
plt.scatter(x["Cholesterol"], x["Age"], c=y_kmeans)
plt.xlabel("Cholesterol")
plt.ylabel("Age")
plt.show()
print("Colors = actual clusters")
plt.scatter(x["Cholesterol"], x["Age"], c=y["Drug"])
plt.xlabel("Cholesterol")
plt.ylabel("Age")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(x[["Na_to_K", "Age"]])
y_kmeans = kmeans.predict(x[["Na_to_K", "Age"]])
print("Colors = predicted clusters")
plt.scatter(x["Na_to_K"], x["Age"], c=y_kmeans)
plt.xlabel("Na_to_K")
plt.ylabel("Age")
plt.show()
print("Colors = actual clusters")
plt.scatter(x["Na_to_K"], x["Age"], c=y["Drug"])
plt.xlabel("Na_to_K")
plt.ylabel("Age")
plt.show()
```
