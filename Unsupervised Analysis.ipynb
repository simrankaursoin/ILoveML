```python
from sklearn import preprocessing
from sklearn.cluster import KMeans
import numpy as np
import math
import pandas as pd
import seaborn

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
drug_df = pd.read_csv("drug200.csv")
drug_df.head()
```

```python
# drug type is the target feature
x = drug_df[["Age", "Sex", "BP", "Cholesterol", "Na_to_K"]]
y = drug_df[["Drug"]]
```

<h1>Visualization</h1>

```python
plt.scatter(y["Drug"], x["Age"], color="blue")
plt.ylabel("Age")
plt.show()
plt.scatter(y["Drug"], x["Sex"], color="red")
plt.ylabel("Sex")
plt.show()
plt.scatter(y["Drug"], x["BP"], color="green")
plt.ylabel("BP")
plt.show()
plt.scatter(y["Drug"], x["Cholesterol"], color="orange")
plt.ylabel("Cholesterol")
plt.show()
plt.scatter(y["Drug"], x["Na_to_K"], color="purple")
plt.ylabel("Na_to_K")
plt.show()
```
